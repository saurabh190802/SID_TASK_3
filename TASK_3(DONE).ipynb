Making a new column called SalePriceClass, based on the following conditions.

if SalePrice greater than or equal 400000 then SalePriceClass = 'A' if SalePrice between 300000 and 399999 then SalePriceClass = 'B' if SalePrice between 200000 and 299999 then SalePriceClass = 'C' if SalePrice between 100000 and 199999 then SalePriceClass = 'D' if SalePrice less than or equal 100000 then SalePriceClass = 'E'

```python
categorical_nan=[feature for feature in final_task.columns if final_task[feature].isnull().sum()>1 and final_task[feature].dtypes=='O']
for feature in categorical_nan:
   print("{}: {}% missing values".format(feature,np.round(final_task[feature].isnull().mean(),4)))
```

REPLACING ALL THE NAN VALUES FOR CATEGORICAL FEATURES.

```python
def replace_cat_feature(data,features_nan):
    dataset=data.copy()
    dataset[categorical_nan]=dataset[categorical_nan].fillna('Missing')
    return dataset

final_task=replace_cat_feature(final_task,categorical_nan)

final_task[categorical_nan].isnull().sum()
```

```python
final_task.head()
```

```python
numerical_with_nan=[feature for feature in final_task.columns if final_task[feature].isnull().sum()>1 and final_task[feature].dtypes!='O']

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.round(final_task[feature].isnull().mean(),4)))
```

## REPLACING THE MISSING VALUES FOR NUMERICAL FEATURES USING MEDIAN & REPLACING IT WITH MEDIAN AS THERE ARE LARGE NUMBER OF OUTLIERS.

```python
for feature in numerical_with_nan:
    median_value=final_task[feature].median()
    
    final_task[feature].fillna(median_value,inplace=True)
    
final_task[numerical_with_nan].isnull().sum()
```

```python
final_task.head()
```

## REPLACING THE CATEGORIES OF EACH FEATURE OCCURING LESS THAN 0.01 BY 'RARE VAR' LABEL

```python
for feature in categorical_features:
    temp=final_task.groupby(feature)['SalePrice'].count()/len(final_task)       #GROUPING BASED ON CATEGORIES OF EACH FEATURE AND COUNTING THEM,THEN DIVIDING THEM BY TOTAL NUMBER OF COLUMNS
    temp_df=temp[temp>0.01].index                                   #STORING THE INDEX WHOSE VALUE IS MORE THAN 0.01
    final_task[feature]=np.where(final_task[feature].isin(temp_df),final_task[feature],'Rare_var')                      #IF VALUE IN TEMP_DF THEN IT WILLREMAIN AS IT IS ELSE WILL BE PEPLACED BY RARE_VAR
```

```python
final_task.head(10)
```

# LABEL ENCODING

```python
for feature in categorical_features:
    labels_ordered=final_task.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    final_task[feature]=final_task[feature].map(labels_ordered)
```
